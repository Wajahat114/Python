```python
import pandas as pd
```

# Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data={'values':[100,200,300,400,500,600,700,800,900,1000]}
df=pd.DataFrame(data)
df.head()
```

```python
scaler=MinMaxScaler()
df['scaled_values']=scaler.fit_transform(df[['values']])
df.head(10)
```

# 2. Standard Scaler or Z-score normalization

```python
from sklearn.preprocessing import StandardScaler
scaler1=StandardScaler()
df['scaled_values1']=scaler1.fit_transform(df[['values']])
```

```python
df.head(10)
```

# 3. Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
scaler2=RobustScaler()
df['scaled_values2']=scaler2.fit_transform(df[['values']])
```

```python
df.head(10)
```

# 4. Log Scaler

```python
import numpy as np
df['scaled_values3']=np.log10(df['values'])
```

```python
df.head(10)
```
