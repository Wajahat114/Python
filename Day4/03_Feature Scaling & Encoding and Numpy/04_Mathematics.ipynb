```python
import numpy as np
```

```python
# coefficient matrix A
A= np.array([[2,3],[5,4]])
print(A)
```

```python
# constant vector b
B= np.array([8,20])
print(B)
```

```python
# solve for x
x=np.linalg.solve(A,B)
print(float(x[0]),float(x[1]))
```

# Solving Complex Equation

```python
A=np.array([[1+2j,2-1j],[3-1j,4]])
B=np.array([4+1j,3+4j])
x=np.linalg.solve(A,B)
x
```

# Vectors in python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a vector
V=np.array([2,3])
V
```

```python
plt.figure()
plt.quiver(0,0,V[0],V[1],angles='xy',scale_units='xy',scale=1)
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.title('Vector V')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

## Transforming a Vector

### Rotation

```python
theta=np.radians(-90)
rm=np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
V=np.array([3,3])
Vt=rm.dot(V)
print(Vt)
# plotting the results
plt.figure()
plt.quiver(0,0,V[0],V[1],angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,Vt[0],Vt[1],angles='xy',scale_units='xy',scale=1,color='b')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid()
#plotting x axis and y axis line
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.title('Vector V')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['V','transformed V'])
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

### Shear Transformation

```python
Shear_factor=1.5
shear_matrix=np.array([[1,Shear_factor],[0,1]])
V=np.array([3,3])
Vt1=shear_matrix.dot(V)
# ploting the results
plt.figure()
plt.quiver(0,0,V[0],V[1],angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,Vt[0],Vt[1],angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,Vt1[0],Vt1[1],angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-5,10)
plt.ylim(-5,10)
plt.grid()
#plotting x axis and y axis line
plt.axhline(y=0,color='k') 
plt.axvline(x=0,color='k')
plt.title('Vector V')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['V','transformed V','sheared V'])
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

```python
# creating shear grid points
x=np.linspace(-5,10,10)
print(x)
y=np.linspace(-5,10,10)
X, Y=np.meshgrid(x, y)
print(X)
```

```python
X_shear, Y_shear=shear_matrix.dot([X.flatten(),Y.flatten()])
X_shear=X_shear.reshape(X.shape)
Y_shear=Y_shear.reshape(Y.shape)
```

```python
# plotting with shear grid
plt.figure()

plt.quiver(0,0,V[0],V[1],angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,Vt[0],Vt[1],angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,Vt1[0],Vt1[1],angles='xy',scale_units='xy',scale=1,color='g')

#plotting shear grid
for i in range(len(x)):
    plt.plot(X_shear[i,:],Y_shear[i,:],'k--',linewidth=0.5)
    plt.plot(X_shear[:,i],Y_shear[:,i],'k--',linewidth=0.5)



plt.xlim(-5,10)
plt.ylim(-5,10)
#plotting x axis and y axis line
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.title('Vector V')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['V','transformed V','sheared V'])
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```
