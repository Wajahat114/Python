```python
df=sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# scatter plot
plt.figure(figsize=(12,8))
sns.scatterplot(x='age',y='fare',data=df, hue='sex', palette='RdYlBu', size='class', sizes=(20,200), alpha=1)
plt.title('Titanic Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
df.groupby("sex")["fare"].max()
```

```python
# numerber of unuque values in fare column
df['fare'].max()
```

```python
# dataset row which has max fare
df[df['fare']==df['fare'].max()]
```
