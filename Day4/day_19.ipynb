# Data Viz

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic dataset from seaborn
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter',x='age',y='fare',color='red')
```

```python
df.plot(kind='scatter',x='age',y='fare',color='#00ffae')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar',xlabel='Gender',ylabel='Mean Age')
```

```python
#seaborn plots
sns.barplot(data=df,x='survived',y='fare')
```

```python
sns.boxplot(data=df,x='sex',y='age',hue='class')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```
