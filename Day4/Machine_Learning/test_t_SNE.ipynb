```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Generate a random dataset with 100 samples
np.random.seed(42)
velocity = np.random.normal(50, 10, 100)  # Normally distributed velocity
temperature = np.random.normal(300, 20, 100)  # Normally distributed temperature
concentration = np.random.uniform(0, 1, 100)  # Uniformly distributed concentration

# Create a DataFrame
df = pd.DataFrame({
    'Velocity': velocity,
    'Temperature': temperature,
    'Concentration': concentration
})

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.values)

# Apply t-SNE (reduce to 2D)
tsne = TSNE(n_components=2, perplexity=10, random_state=42, init='pca')
X_embedded = tsne.fit_transform(X_scaled)

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=df["Velocity"], cmap='coolwarm', alpha=0.7)
plt.colorbar(label="Velocity")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE Visualization of Random Data (100 Samples)")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Generate a random dataset with 100 samples
np.random.seed(42)
velocity = np.random.normal(50, 10, 100)  # Normally distributed velocity
temperature = np.random.normal(300, 20, 100)  # Normally distributed temperature
concentration = np.random.uniform(0, 1, 100)  # Uniformly distributed concentration

# Create a DataFrame
df = pd.DataFrame({
    'Velocity': velocity,
    'Temperature': temperature,
    'Concentration': concentration
})

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.values)

# Apply t-SNE (reduce to 2D)
tsne = TSNE(n_components=2, perplexity=10, random_state=42, init='pca')
X_embedded = tsne.fit_transform(X_scaled)

# Create subplots for each feature
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
features = ['Velocity', 'Temperature', 'Concentration']
colormaps = ['coolwarm', 'viridis', 'plasma']

for i, (feature, cmap) in enumerate(zip(features, colormaps)):
    scatter = axes[i].scatter(X_embedded[:, 0], X_embedded[:, 1], c=df[feature], cmap=cmap, alpha=0.7)
    fig.colorbar(scatter, ax=axes[i], label=feature)
    axes[i].set_xlabel("t-SNE Component 1")
    axes[i].set_ylabel("t-SNE Component 2")
    axes[i].set_title(f"t-SNE Colored by {feature}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Generate a random dataset with 100 samples
np.random.seed(42)
velocity = np.random.normal(50, 10, 100)  # Normally distributed velocity
temperature = np.random.normal(300, 20, 100)  # Normally distributed temperature
concentration = np.random.uniform(0, 1, 100)  # Uniformly distributed concentration

# Create a DataFrame
df = pd.DataFrame({
    'Velocity': velocity,
    'Temperature': temperature,
    'Concentration': concentration
})

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.values)

# Apply t-SNE (reduce to 2D)
tsne = TSNE(n_components=2, perplexity=80, random_state=42, init='pca')
X_embedded = tsne.fit_transform(X_scaled)

# Apply K-Means clustering to create 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Plot the clustered data
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=clusters, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label="Cluster")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE Visualization with 3 Clusters (Velocity, Temperature, Concentration)")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Generate a random dataset with 100 samples
np.random.seed(42)
velocity = np.random.normal(50, 10, 100)  # Normally distributed velocity
temperature = np.random.normal(300, 20, 100)  # Normally distributed temperature
concentration = np.random.uniform(0, 1, 100)  # Uniformly distributed concentration

# Create a DataFrame
df = pd.DataFrame({
    'Velocity': velocity,
    'Temperature': temperature,
    'Concentration': concentration
})

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.values)

# Apply t-SNE (reduce to 2D)
tsne = TSNE(n_components=2, perplexity=10, random_state=42, init='pca')
X_embedded = tsne.fit_transform(X_scaled)

# Apply K-Means clustering to create 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Assign cluster labels
labels = ['Velocity', 'Temperature', 'Concentration']
colors = ['r', 'g', 'b']

# Plot the clustered data
plt.figure(figsize=(8, 6))
for i in range(3):
    plt.scatter(X_embedded[clusters == i, 0], X_embedded[clusters == i, 1], 
                c=colors[i], label=labels[i], alpha=0.7)

plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE Visualization with 3 Clusters (Velocity, Temperature, Concentration)")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Generate a random dataset with 100 samples
np.random.seed(42)
velocity = np.random.normal(50, 10, 100)  # Normally distributed velocity
temperature = np.random.normal(300, 20, 100)  # Normally distributed temperature
concentration = np.random.uniform(0, 1, 100)  # Uniformly distributed concentration

# Create a DataFrame
df = pd.DataFrame({
    'Velocity': velocity,
    'Temperature': temperature,
    'Concentration': concentration
})

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.values)

# Apply t-SNE (reduce to 2D)
tsne = TSNE(n_components=2, perplexity=10, random_state=42, init='pca')
X_embedded = tsne.fit_transform(X_scaled)

# Apply K-Means clustering to create 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Assign colors for each feature
def assign_feature_color(feature_values):
    return np.digitize(feature_values, bins=np.percentile(feature_values, [33, 66]))

# Assign cluster labels
labels = ['Velocity', 'Temperature', 'Concentration']
colors = ['r', 'g', 'b']

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot actual feature clusters using t-SNE components, based on velocity
feature_colors = assign_feature_color(df['Velocity'])
for i in range(3):
    axes[0].scatter(X_embedded[feature_colors == i, 0], X_embedded[feature_colors == i, 1], 
                    c=colors[i], label=labels[i], alpha=0.7)
axes[0].set_xlabel("t-SNE Component 1")
axes[0].set_ylabel("t-SNE Component 2")
axes[0].set_title("Actual Clusters Based on Velocity")
axes[0].legend()

# Plot t-SNE clustered data
for i in range(3):
    axes[1].scatter(X_embedded[clusters == i, 0], X_embedded[clusters == i, 1], 
                    c=colors[i], label=labels[i], alpha=0.7)
axes[1].set_xlabel("t-SNE Component 1")
axes[1].set_ylabel("t-SNE Component 2")
axes[1].set_title("t-SNE Clustering Visualization")
axes[1].legend()

plt.tight_layout()
plt.show()
```
